```python
%matplotlib notebook
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
f=pd.read_csv('Fruits_data.csv')
```

```python
y=f.drop('fruit_label',axis='columns');
```

```python
target=f['fruit_label']
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
nu_fruit_name=LabelEncoder()
nu_fruit_subtype=LabelEncoder()
y['Fruit_name_num']=nu_fruit_name.fit_transform(y['fruit_name'])
y['Fruit_subtype_num']=nu_fruit_subtype.fit_transform(y['fruit_subtype'])
y
```

```python
z_feature=y.drop(['fruit_name','fruit_subtype','width','Unnamed: 0',
                  'Fruit_subtype_num','color_score'],axis='columns')
```

```python
x_train,x_test,y_train,y_test=train_test_split(z_feature,target)
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig=plt.figure()
ax=fig.add_subplot(projection='3d')
ax.scatter(x_train['mass'],x_train['Fruit_name_num'],x_train['height'],c=y_train,marker='o',s=100)
ax.set_xlabel('mass')
ax.set_ylabel('Fruit_name_num')
ax.set_zlabel('height')
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier(n_neighbors=7)
knn.fit(x_train,y_train)
```

```python
knn.score(x_test,y_test)
```
